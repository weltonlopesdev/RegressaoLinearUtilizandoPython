Modelo que faz relação entre as variaveis X e Y **(variavel dependente Y) (variavel explanatória X)**

> Velocidade (y)

y = previsão do tempo
x = temperatura
B0(beta 0) = Constante
B1(beta 1) = coeficiente

image.png

B0 e B1 = definem a localização da linha (treinamento)



> B0 = é a RETA

> B1 = é o que intercepta o X1 e B0






image.png

ERRO: É o espaço entre a reta e os pontos ou seja o (e)

**Sendo assim temos a formula :** y = B0 + B1*x1 + e

**variável preditora:** é a variável independente, que tem o poder de influenciar na variável que nós queremos encontrar

**variável alvo ou dependente:** é a variável que queremos prever. 

```python
# importando as libs
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# variável preditora
X = np.array([ 220, 220, 220, 220, 220, 225, 225, 225, 225, 225, 230, 230, 230, 230, 230, 235, 235, 235, 235, 235 ])
```

```python
# variável alvo
y = np.array([ 137, 137, 137, 136, 135, 135, 133, 132, 133, 133, 128, 124, 126, 129, 126, 122, 122, 122, 119, 122 ])
```

```python
# é necessário adicionar uma constante a matriz X
X_sm = sm.add_constant(X)
# OLS vem de Ordinary Least Squares e o método fit irá treinar o modelo
results = sm.OLS(y, X_sm).fit()
# mostrando as estatísticas do modelo
results.summary()
# mostrando as previsões para o mesmo conjunto passado
results.predict(X_sm)
```

```python
df = pd.DataFrame()
df['x'] = X
df['y'] = y
# passando os valores de x e y como Dataframes
x_v = df[['x']]
y_v = df[['y']]
# criando e treinando o modelo
model = LinearRegression()
model.fit(x_v, y_v)
# para visualizar os coeficientes encontrados
model.coef_
```
